```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, SimpleRNN, Dense

# Define input shape
timestep = 10     # number of time steps in each sequence
features = 5      # number of features per time step

# Build the model
model = Sequential([
    LSTM(100, activation='tanh', return_sequences=True, input_shape=(timestep, features)),
    GRU(60, activation='tanh', return_sequences=True),
    SimpleRNN(40, activation='tanh'),
    Dense(1, activation='sigmoid')   # Sigmoid for binary classification
])

# Compile the model
model.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

# Show model summary
model.summary()
```
