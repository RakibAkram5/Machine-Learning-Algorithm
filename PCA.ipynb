```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
X=np.array([
    [2,1],
    [4,3],
    [6,5],
    [8,7] 
])
pca=PCA(n_components=1)
z=pca.fit_transform(X)
print("Reduce Data: " )
print(z)
print("\n Eigenvalues:")
print(pca.explained_variance_)
print("\n Principal Components: ")
print(pca.components_)
```
